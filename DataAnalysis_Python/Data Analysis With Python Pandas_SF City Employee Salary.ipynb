```python
#6. Drop ID, Notes, Agency, and Status Columns. axis=0 to drop rows and 1 to col.
data.columns
```

```python
data.drop(['Id','Notes','Agency','Status'],axis=1)
```

```python
#7. Get Overall Statistics About The Dataframe
data.describe(include='all')
```

```python
#8. Find Occurrence of The Employee Names  (Top 5)
data['EmployeeName'].value_counts().head(5)
```

```python
#9. Find The Number of Unique Job Titles
len(data['JobTitle'].value_counts())
data['JobTitle'].nunique()
```

```python
#10.Total Number of Job Titles Contain Captain
data[data['JobTitle'].str.contains('Captain')]
```

```python
data[data['JobTitle'].str.contains('Captain',case=False)]
```

```python
len(data[data['JobTitle'].str.contains('Captain')])
```

```python
#11. Display All the Employee Names From Fire Department, also show JobTitle
data[data['JobTitle'].str.contains('Fire',case=False)][['EmployeeName','JobTitle']]
```

```python
#12. Find Minimum, Maximum, and Average BasePay
data['BasePay'].describe(include='all')
```

```python
#BasePay has NAN
data.describe(include='all')
```

```python
#13. Replace 'Not Provided' in EmployeeName' Column to NaN 
import numpy as np
#data[data['EmployeeName'].str.contains('Not Provided',case=False)]
```

```python
data['EmployeeName']=data['EmployeeName'].replace('Not provided',np.nan)
```

```python
data['EmployeeName']
```

```python
#14. Drop The Rows Having 5 Missing Values
data.isnull().sum()
#above output is across columns, we need rows having 5 missing  values. 0 records for 5 missig values, so took 3
data[data.isnull().sum(axis=1)==5]
```

```python
data.drop(data[data.isnull().sum(axis=1)==5].index,axis=0)
```

```python
#15. Find Job Title of ALBERT PARDINI
data[data['EmployeeName']=='ALBERT PARDINI']['JobTitle']
```

```python
#16. How Much ALBERT PARDINI Make (Include Benefits)?
data[data['EmployeeName']=='ALBERT PARDINI']['TotalPayBenefits']
```

```python
#17.Display Name of The Person Having The Highest BasePay / TotalPay
data[data['TotalPay'].max()==data['TotalPay']]['EmployeeName']
```

```python
#18.Find Average BasePay of All Employee Per Year 
data.groupby('Year').mean()['TotalPay']
```

```python
#19. Find Average BasePay of All Employee Per JobTitle 
data.groupby('JobTitle').mean()['TotalPay']
```

```python
#20. Find Average BasePay of Employee Having Job Title ACCOUNTANT  
data[data['JobTitle']=='ACCOUNTANT'].mean() ['BasePay']
```

```python
data.groupby('JobTitle').mean()['TotalPay']
```

```python
#21. Find Top 5 Most Common Jobs
data['JobTitle'].value_counts().head(5)
```
